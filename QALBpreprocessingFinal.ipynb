# Removing punctuations & numbers

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, 'NaN')
    return text
```

```python
def remove_numbers(text):
    for number in string.digits:
        text = text.replace(number, 'NaN')
    return text    
```

```python
data['word_without_tahskeel'] = data['word_without_tahskeel'].apply(remove_punctuations)
data['word_without_tahskeel'] = data['word_without_tahskeel'].apply(remove_numbers)
```

```python
deleteRow = data[ data['word_without_tahskeel'] == 'NaN' ].index
data.drop(deleteRow , inplace=True)
```

# Remove Stop Words

```python
#nltk.download("stopwords")
ar_stopwords = set(nltk.corpus.stopwords.words("arabic"))
print(ar_stopwords)
```

```python
def removing_ar_stopwords(text):
    for word in ar_stopwords:
        if text == word:
            text = text.replace(word, 'NaN')
    return text   
```

```python
data['word_without_tahskeel'] = data['word_without_tahskeel'].apply(removing_ar_stopwords)
data.head(30)
```

```python
deleteStopWords = data[ data['word_without_tahskeel'] == 'NaN' ].index
data.drop(deleteStopWords , inplace=True)
data.head(30)
```

# Normalization

```python
import re
def normalizeArabic(text):
    text = text.strip()
    text = re.sub("[إأٱآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "ء", text)
    text = re.sub("ئ", "ء", text)
    text= re.sub("ة", "ه", text)
    text = re.sub(r'(.)\1+', r"\1\1", text) # Remove longation
    return(text)
```

```python
data['word_without_tahskeel'] = data['word_without_tahskeel'].apply(normalizeArabic)
data.head(30)
```

# Lemmatization

```python
from nltk.stem.isri import ISRIStemmer
def lemmatization(text):
    stemWord = ISRIStemmer().suf32(text)
    return stemWord
  
```

```python
data['word_without_tahskeel'] = data['word_without_tahskeel'].apply(lemmatization)
data.head(30)
```
